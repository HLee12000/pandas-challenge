```python
import pandas as pd
from pathlib import Path

#Pathing to files
school_complete_data = Path("Resources/schools_complete.csv")
student_complete_data = Path("Resources/students_complete.csv")

#Reading files in
school_data = pd.read_csv(school_complete_data)
student_data = pd.read_csv(student_complete_data)
```

```python
#Merging data sets
school_student_complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

```python
#Calc unique schools
school_names = len(school_student_complete_data["school_name"].unique())
```

```python
#Calc total students
total_students = school_student_complete_data["student_name"].count()
```

```python
#Calc total budget
total_budget = sum(school_student_complete_data["budget"].unique())
```

```python
#Calc avg math score
average_math_score = school_student_complete_data["math_score"].mean()
```

```python
#Calc avg reading score
average_reading_score = school_student_complete_data["reading_score"].mean()
```

```python
#Count students passing math(>=70) and calc percentage passing 
passing_math_count = school_student_complete_data[((school_student_complete_data["math_score"])>=70)].count()["student_name"]
passing_math_percentage = passing_math_count / total_students *100
```

```python
#Count students passing reading(>=70) and calc percentage passing 
passing_reading_count = school_student_complete_data[((school_student_complete_data["reading_score"])>=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / total_students *100
```

```python
#Count students passing reading and math(>=70) and calc percentage passing both 
overall_passing_count = school_student_complete_data[(school_student_complete_data["math_score"]>=70) & (school_student_complete_data["reading_score"] >= 70)].count()["student_name"]
overall_passing_percentage = overall_passing_count / total_students *100
```

```python
#Creating dataframe for district.
District_data_summary = pd.DataFrame({"Total Schools":[school_names],
                                     "Total Students":[total_students],
                                     "Total Budget":[total_budget],
                                     "Average Math Score":[average_math_score],
                                     "Average Reading Score":[average_reading_score],
                                     "% Passing Math":[passing_math_percentage],
                                     "% Passing Reading":[passing_reading_percentage],
                                     "% Overall Passing":[overall_passing_percentage],
                                     })
#This code can be used to fomat certain numbers
District_data_summary["Total Students"] = District_data_summary["Total Students"].map("{:,}".format)
District_data_summary["Total Budget"] = District_data_summary["Total Budget"].map("${:,.2f}".format)
District_data_summary
```

# School Summary

```python
#Indexing schoolname with type
school_sum_types = school_data.set_index(["school_name"])["type"]
```

```python
#Indexing schoolname with size
per_school_sum_counts = school_data.set_index(["school_name"])["size"]
```

```python
#Calculating budget for each school and spending per capita(students)
per_school_sum_budget = school_data.set_index(["school_name"])["budget"]
school_data["per_school_capita"] = school_data["budget"] / school_data ["size"]
per_school_capita = school_data.set_index(["school_name"])['per_school_capita']
```

```python
#Calc average math and reading score
school_sum_math_avg = school_student_complete_data.groupby("school_name")["math_score"].mean()
school_data["school_sum_math_avg"] = school_sum_math_avg
school_sum_reading_avg = school_student_complete_data.groupby("school_name")["reading_score"].mean()
school_data["school_sum_reading_avg"] = school_sum_reading_avg
```

```python
#counting students passing math by school
passing_math_students = school_student_complete_data[school_student_complete_data['math_score']>=70]
school_sum_math_pass_count = passing_math_students.groupby('school_name').size()
```

```python
#counting students passing reading by school
passing_reading_students = school_student_complete_data[school_student_complete_data['reading_score']>=70]
school_sum_reading_pass_count = passing_reading_students.groupby('school_name').size()
```

```python
#counting students passing math and reading by school
passing_math_reading_students = school_student_complete_data[(school_student_complete_data["reading_score"]>=70)&(school_student_complete_data["math_score"]>=70)]
school_passing_math_reading = passing_math_reading_students.groupby(["school_name"]).size()
```

```python
#calculating percent for above 3 blocks of code
per_school_sum_passing_math = school_sum_math_pass_count / per_school_sum_counts * 100
per_school_sum_passing_reading = school_sum_reading_pass_count / per_school_sum_counts * 100
per_school_sum_passing_math_and_reading = school_passing_math_reading / per_school_sum_counts * 100
```

```python
#creating data frame for per school summary
per_school_summary = pd.DataFrame({'Type':school_sum_types,
                           'Total Students':per_school_sum_counts,
                           'Total Budget':per_school_sum_budget,
                           'Per Student Budget':per_school_capita,        
                           'Average Math Score':school_sum_math_avg,
                           'Average Reading Score':school_sum_reading_avg,
                           '% Passing Math':per_school_sum_passing_math,
                           '% Passing Reading':per_school_sum_passing_reading,
                           '% Overall Passing':per_school_sum_passing_math_and_reading,
                                     })

per_school_summary.index.name = None
#CAUTION: Below code can be used once for formatting, however will break code further down because it will try to
#do calculations on strings. This code is for assignment sake, will not run as to not break code
##per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
##per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary
```

# Scores by School Spending

```python
#create bins and lavels
spending_bins = [0,585,630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Example of where code will break for previously stated error. creating copy 
school_spending_df = per_school_summary.copy()
```

```python
#pd.cut the Per student budget by the spending bins listed above with ascribed labels 
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'],
                                                            bins=spending_bins, labels=labels)
```

```python
#calculations
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
#dataframe for above 
spending_summary = pd.DataFrame ({"Average Math Score": spending_math_scores,
                                  "Average Reading Score": spending_reading_scores,
                                  "% Passing Math": spending_passing_math,
                                  "% Passing Reading": spending_passing_reading,
                                  "% Overall Passing": overall_passing_spending
})
spending_summary
```

# Scores by School Size

```python
#create bins and lavels
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#pd.cut the per school size (total students) by the size bins listed above with ascribed labels 
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"],bins=size_bins, labels=labels)
```

```python
#calculations
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
#dataframe for above 
size_summary = pd.DataFrame({"Average Math Score": size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Math": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing": size_overall_passing
})
size_summary
```
